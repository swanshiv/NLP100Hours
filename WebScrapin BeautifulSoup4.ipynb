```python
import requests
```

```python
web_page = requests.get("https://www.weather-forecast.com/locations/Bangalore/forecasts/latest")
```

```python
web_page.content
```

```python
from bs4 import BeautifulSoup
soup = BeautifulSoup(web_page.content, 'html.parser')
```

```python
print(soup.prettify())
```

```python
title = soup.title
```

```python
print(title)
```

```python
title = soup.title.text
print(title)
```

```python
table = soup.find('table')
table
```

```python
# Find the table rows within the table
table_rows = table.find_all('tr')
```

```python
# We can iterate through the rows, find the 'td' tags and print out each of the table data
for tr in table_rows:
    td = tr.find_all('td')
    row = [i.text for i in td]
    print(row)
```
